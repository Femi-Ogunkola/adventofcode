```python
f = open("AOC6.txt", "r")
test = (f.read().split('\n'))
print(test)
```

```python
for i in range(len(test)):
    test[i] = test[i].replace('\n',',')
print(test)
```

```python
def nSplit(lst, delim, count=1):
    output = [[]]
    delimCount = 0
    for item in lst:
        if item == delim:
            delimCount += 1
        elif delimCount >= count:
            output.append([item])
            delimCount = 0
        else:
            output[-1].append(item)
            delimCount = 0
    return output

        
```

```python
new = nSplit(test,'')
print(new)
```

```python
def setter(check):
    for i in range(len(check)):
        check[i] = set(check[i])

    return(len(set.intersection(*check)))
    
```

```python
count = 0
new = nSplit(test,'')
for i in range(len(new)):
    count=count+(setter(new[i]))

print(count)
```
