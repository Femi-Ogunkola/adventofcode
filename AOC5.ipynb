```python
def split(arr,c):
    if c =='F':
        new_len_a = len(arr)//2
        arr = arr[:new_len_a]
    if c =='B':
        new_len_a = len(arr)//2
        arr = arr[new_len_a:]
    return arr

def c_split(r,c):
    if c =='R':
        new_len_a = len(r)//2
        r = r[new_len_a:]
    if c =='L':
        new_len_a = len(r)//2
        r = r[:new_len_a]
    return r
```

```python
def get_max(a):
    row = list(range(0, 128))
    column = list(range(0,8))

    for i in range(len(a)-3):
        row = split(row,a[i])
 

    for i in range(len(a[-3:])):
        column = c_split(column,a[i+7])

     

    return(row[0]*8+column[0])  
```

```python
 p = []
for i in final:
    p.append(get_max(i))
```
